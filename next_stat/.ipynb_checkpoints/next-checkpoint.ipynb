```python
%matplotlib inline
import requests
from bs4 import BeautifulSoup as bs
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import sys
from tqdm import tqdm
matplotlib.rc('xtick', labelsize=12) 
matplotlib.rc('ytick', labelsize=12) 
```

## 从GCN网站中读取含指定关键字的标题及标题的内容

```python
def get_gcn(url, keyword):
    """
    find entries whose title includes given keyword
    
    params:
    ---
    url: str
        gcn url
        
    keyword: str
        keyword
    
    returns:
    ---
    infodict: dict
        containing: index, single url, title, content
    """
    
    re=requests.get(url)
    soup=bs(re.text, features='html.parser')
    sys.setrecursionlimit(10000000)
    lis=soup.find_all('li')
    
    info={}
    
    entries=str(lis[8]).split('\n')[:-1]
    bs(entries[0], 'html.parser').find('li').contents[-2].strip(' ')
    

    for entry in tqdm(entries):
        bss=bs(entry,'html.parser')
        if bss.find('a') == None: continue
        gcnurl='https://gcn.gsfc.nasa.gov/'+bss.find('a')['href']
        idx=bss.find('a').string
        title=bss.find('li').contents[-2].strip(' ')
        if keyword in title:
            try:
                content=get_gcn_content(gcnurl)
            except:
                content='not found.'
            info[idx]=[title, gcnurl, content]
        
    return info
```

```python
def get_gcn_content(url):
    """
    get contents of a gcn notice with given url
    
    Param
    ---
    url: str
    
    Return
    ---
    content: str
    """
    
    re=requests.get(url)
    content=re.text.split('report')[-1].strip(': \n').replace('\n', '', 100)
    
    return content
```

## processes

```python
url='https://gcn.gsfc.nasa.gov/gcn3_archive.html'
re=requests.get(url)
soup=bs(re.text, features='html.parser')
sys.setrecursionlimit(10000000)
lis=soup.find_all('li')
gcnlist=[] # GCN INDEX
titlelist=[] # gcn title
urllist=[] # url for individual gcn notice
contentls=[] # content of a gcn notice, via urllist
grbname=[]
entries=str(lis[8]).split('\n')[:-1]
bs(entries[0], 'html.parser').find('li').contents[-2].strip(' ')

for entry in tqdm(entries):
    bss=bs(entry,'html.parser')
    if bss.find('a') == None: continue
    gcnurl='https://gcn.gsfc.nasa.gov/'+bss.find('a')['href']
    idx=bss.find('a').string
    title=bss.find('li').contents[-2].strip(' ')
    if 'NEXT' in title and 'GRB' in title: 
        grbname.append(title.split(':')[0])
        gcnlist.append(idx)
        titlelist.append(title)
        urllist.append(gcnurl)
        contentls.append(get_gcn_content(gcnurl))

with open('next_gcngrb.csv','w') as log:
    for i in range(len(gcnlist)):
        gcnidx=gcnlist[i]
        gcntt=titlelist[i]
        grbnm=grbname[i]
        noticeurl=urllist[i]
        content=contentls[i]
        log.write('{}|{}|{}|{}|{}\n'.format(gcnidx, gcntt, grbnm, noticeurl, content))
```

## 获取完毕, 接下来获取T90和红移. 手动获取完毕

```python
import pandas as pd
```

```python
nts=pd.read_csv('./next_auto.csv')
b,c=nts.iloc[10].copy(),nts.iloc[11].copy()
nts.iloc[10], nts.iloc[11]=c,b
nts=nts.drop(5)
```

```python
nts
```

```python
grbname=list(nts['name'])[::-1]
t90=list(map(float, nts['t90']))[::-1]
tobs=[float(str(x).split('(')[0]) for x in nts['tobs']][::-1]
mag=[float(str(y).split('(')[0]) for y in nts['mag']][::-1]
z=[]
for rs in nts['z']:
    reds = float(rs.split('(')[0])
    if reds == 999: reds=0
    z.append(reds)
z=z[::-1]
```

```python
x=list(range(1,len(grbname)+1))
```

```python
fig, axs = plt.subplots(4,1,sharex=True, figsize=(20,10), constrained_layout=False) 
# constrained_layout automatically adjusts subplots and decorations like legends and colorbars so that they fit in the figure window while still preserving, as best they can, the logical layout requested by the user.
fig.subplots_adjust(hspace=0.1)
fig.suptitle('GRBs automatically detected by NEXT', fontsize=18, y=0.93)

axs[0].scatter(x, t90, marker='x')
axs[0].set_xticks(x)
axs[0].set_xticklabels([name.split(' ')[-1] for name in grbname])
axs[0].set_ylabel('t90', fontsize=15)
axs[0].grid(True, linestyle='-.')

axs[1].scatter(x, tobs, marker='x', color='g')
axs[1].set_xticks(x)
axs[1].set_xticklabels([name.split(' ')[-1] for name in grbname])
axs[1].set_ylabel('tobs', fontsize=15)
axs[1].grid(True, linestyle='-.')

axs[2].scatter(x, mag, marker='x', color='m')
axs[2].set_xticks(x)
axs[2].set_xticklabels([name.split(' ')[-1] for name in grbname])
axs[2].set_ylabel('mag', fontsize=15)
axs[2].grid(True, linestyle='-.')

axs[3].scatter(x, z, marker='x', color='r')
axs[3].set_xticks(x)
axs[3].set_xticklabels([name.split(' ')[-1] for name in grbname])
axs[3].set_ylabel('redshift', fontsize=15)
axs[3].grid(True, linestyle='-.')

for ax in axs.flat:
    ax.label_outer()

fig.savefig('next_auto.jpg',bbox_inches='tight', dpi=300)
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec


def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        #ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
        #ax.tick_params(labelbottom=False, labelleft=False)
        pass

fig = plt.figure(constrained_layout=True, figsize=(22,10))

gs = GridSpec(4, 2, figure=fig)

ax1 = fig.add_subplot(gs[0, 0])
ax1.scatter(x, t90, marker='x')
ax1.set_xticks(x)
ax1.set_xticklabels([name.split(' ')[-1] for name in grbname])
ax1.set_ylabel('t90', fontsize=15)
ax1.grid(True, linestyle='-.')
ax1.tick_params(labelbottom=False, labelleft=True)


ax2 = fig.add_subplot(gs[1,0])
ax2.scatter(x, tobs, marker='x', color='g')
ax2.set_xticks(x)
ax2.set_xticklabels([name.split(' ')[-1] for name in grbname])
ax2.set_ylabel('tobs', fontsize=15)
ax2.grid(True, linestyle='-.')
ax2.tick_params(labelbottom=False, labelleft=True)

ax3 = fig.add_subplot(gs[2,0])
ax3.scatter(x, mag, marker='x', color='m')
ax3.set_xticks(x)
ax3.set_xticklabels([name.split(' ')[-1] for name in grbname])
ax3.set_ylabel('mag', fontsize=15)
ax3.grid(True, linestyle='-.')
ax3.tick_params(labelbottom=False, labelleft=True)

ax4 = fig.add_subplot(gs[3,0])
ax4.scatter(x, z, marker='x', color='r')
ax4.set_xticks(x)
ax4.set_xticklabels([name.split(' ')[-1] for name in grbname])
for tick in ax4.get_xticklabels():  # 将横坐标倾斜30度，纵坐标可用相同方法
    tick.set_rotation(45)
ax4.set_ylabel('redshift', fontsize=15)
ax4.grid(True, linestyle='-.')
ax4.tick_params(labelbottom=True, labelleft=True)

ax5 = fig.add_subplot(gs[:,1])
ax5.scatter(tobs, t90, color='orange')
ax5.set_xlabel('tobs',fontsize=15)
ax5.set_ylabel('t90', fontsize=15)
#ax2.set_ylim(0,1000)
#ax2.set_xlim(0,1000)
ax5.plot(np.linspace(1,500,500),np.linspace(1,500,500), linestyle='--', color='k', label='t90=tobs')
ax5.legend(prop={'size':13})

#fig.suptitle("GridSpec")
format_axes(fig)

plt.suptitle('GRBs automatically observed by NEXT',fontsize=18)
plt.show()
fig.savefig('next_auto.jpg',bbox_inches='tight',dpi=300)
```
